## 1. Imports & LibrairiesOn charge ici toutes les librairies n√©cessaires pour la manipulation de donn√©es, la visualisation et le machine learning.

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
```

## 2. Exploration Visuelle des VariablesCette section permet de visualiser les relations potentielles entre certaines variables candidates et la variable cible √† l‚Äôaide de scatter plots.

```python
DATA_PATH = r"C:\Users\abenech\Downloads\donnees_hopital.csv"

df = pd.read_csv(DATA_PATH, sep=";", encoding="utf-8")
print(df.shape)
df.head()
```

## 3. Pr√©paration du Mod√®leOn r√©importe ici certaines librairies ML afin de construire les pipelines et mod√®les.

```python
display(df.info())

# Manquants par colonne
missing = df.isna().sum().sort_values(ascending=False)
display(missing[missing > 0])

# Doublons
print("Doublons:", df.duplicated().sum())

# Aper√ßu stats num√©riques
display(df.describe(include="number").T)
```

## 4. D√©finition de la Variable CibleOn isole la variable que l‚Äôon souhaite pr√©dire : le nombre total d‚Äôadmissions.

```python
df["date"] = pd.to_datetime(df["date"], format="%d/%m/%Y", errors="coerce")

# Si jamais des dates ont √©chou√©
print("Dates invalides:", df["date"].isna().sum())

df = df.sort_values("date").reset_index(drop=True)
df.head()
```

## 5. Fonction d‚ÄôEntra√Ænement & √âvaluationFonction utilitaire qui pr√©pare les donn√©es, entra√Æne un Random Forest et retourne les m√©triques de performance.

```python
# Cat√©gorielles (selon ton CSV)
cat_cols = ["day_of_week", "season"]
for c in cat_cols:
    if c in df.columns:
        df[c] = df[c].astype("category")

# Bool√©ens (0/1) -> int
bool_like = [
    "heatwave_event", "accident_event", "supply_delivery_day", "it_system_outage"
]
for c in bool_like:
    if c in df.columns:
        df[c] = df[c].astype("Int64")  # accepte NA

df.dtypes
```

## 6. Mod√®le 1 ‚Äì Toutes les Variables (Full Model)Ce mod√®le utilise presque toutes les colonnes disponibles. Il sert de r√©f√©rence haute performance mais peut √™tre irr√©aliste en production.

```python
df["year"] = df["date"].dt.year
df["month"] = df["date"].dt.month
df["day"] = df["date"].dt.day
df["dow"] = df["date"].dt.dayofweek  # lundi=0
df["week"] = df["date"].dt.isocalendar().week.astype(int)
df["is_weekend"] = (df["dow"] >= 5).astype(int)

df[["date","year","month","dow","week","is_weekend"]].head()
```

## 7. Mod√®le 2 ‚Äì Variables Fortement Corr√©l√©esOn s√©lectionne uniquement les variables les plus corr√©l√©es avec la cible pour r√©duire la complexit√© tout en gardant de bonnes performances.

```python
TARGET = "total_admissions"

assert TARGET in df.columns, f"Colonne cible '{TARGET}' introuvable."
display(df[[ "date", TARGET ]].head())
```

## 8. Mod√®le 3 ‚Äì Variables R√©alistesIci on simule un cas r√©el avec uniquement des variables disponibles √† l‚Äôavance. Les performances baissent mais le mod√®le est plus cr√©dible en situation r√©elle.

```python
plt.figure()
plt.plot(df["date"], df[TARGET])
plt.title("Arriv√©es totales (total_admissions) dans le temps")
plt.xlabel("Date")
plt.ylabel("Arriv√©es")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## 9. Comparaison des R√©sultatsTableau r√©capitulatif des performances des diff√©rents mod√®les afin d‚Äôidentifier le meilleur compromis pr√©cision / r√©alisme.

```python
window = 7  # 7 jours
rolling = df[TARGET].rolling(window=window).mean()

plt.figure()
plt.plot(df["date"], df[TARGET], alpha=0.5, label="R√©el")
plt.plot(df["date"], rolling, label=f"Moyenne mobile {window}j")
plt.title("Arriv√©es : r√©el vs tendance (moyenne mobile)")
plt.xlabel("Date")
plt.ylabel("Arriv√©es")
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.figure()
plt.hist(df[TARGET].dropna(), bins=30)
plt.title("Distribution de total_admissions")
plt.xlabel("Arriv√©es")
plt.ylabel("Fr√©quence")
plt.tight_layout()
plt.show()


```
